```python
import yfinance as yf
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
#import ta

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers as lyrs
from tensorflow.python.keras import backend as k

from keras.layers.core import Dense, Activation, Dropout
from keras.layers.recurrent import LSTM
from tensorflow.keras.models import Sequential, load_model

import math, time #helper libraries
import datetime

import itertools
from sklearn import preprocessing
from operator import itemgetter
from sklearn.metrics import mean_squared_error
from math import sqrt
```

# Helper functions

```python
## adapted from: 
## https://github.com/etai83/lstm_stock_prediction/blob/master/.ipynb_checkpoints/GOOGLE%20stock%20prediction-checkpoint.ipynb

def split_data(stock, seq_len):
    num_features = len(stock.columns)
    data = stock.as_matrix() 
    sequence_length = seq_len + 1
    label = []
    
    for index in range(len(data) - sequence_length):
        label.append(data[index: index + sequence_length])

    label = np.array(label)
    row = round(0.9 * label.shape[0])
    train = label[:int(row), :]
    
    x_train = train[:, :-1]
    y_train = train[:, -1][:,-1]
    x_test = label[int(row):, :-1]
    y_test = label[int(row):, -1][:,-1]

    x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1], num_features))
    x_test = np.reshape(x_test, (x_test.shape[0], x_test.shape[1], num_features))  

    return [x_train, y_train, x_test, y_test]
```

```python
def make_model(layers):
        d = 0.2
        model = tf.keras.Sequential()
        model.add(lyrs.LSTM(128, input_shape=(layers[1], layers[0]), return_sequences=True))
        model.add(lyrs.Dropout(d))
        model.add(lyrs.LSTM(64, input_shape=(layers[1], layers[0]), return_sequences=False))
        model.add(lyrs.Dropout(d))
        model.add(lyrs.Dense(16,activation='tanh'))        
        model.add(lyrs.Dense(1,activation='tanh')) 
        model.compile(loss='mse',optimizer='adam',metrics=['accuracy'])
        return model
```

# Data manipulation and Learning

```python
GOOGL = yf.download('GOOGL','2015-06-15','2017-06-16') 
GOOGL2 = yf.download('GOOGL','2010-06-15','2020-06-16') 
ADBE = yf.download('ADBE','2015-06-15','2019-06-16') 
DOX = yf.download('DOX','2015-06-15','2019-06-16') 
```

### Why does FFT spectrum look so different as time scale changes? How to quantify what's going on?

```python
fft = tf.signal.rfft(GOOGL['High'])
f_per_dataset = np.arange(0, len(fft))

n_samples_h = len(GOOGL['High'])
hours_per_year = 24*365.2524
years_per_dataset = n_samples_h/(hours_per_year)

f_per_year = f_per_dataset/years_per_dataset
plt.step(f_per_year, np.abs(fft))
plt.xscale('log')
plt.xlim([0.1, max(plt.xlim())])
plt.xticks([1, 365.2524], labels=['1/Year', '1/day'])
_ = plt.xlabel('Frequency (log scale)')
```

```python
fft = tf.signal.rfft(ADBE['Low'])
f_per_dataset = np.arange(0, len(fft))

n_samples_h = len(GOOGL['High'])
hours_per_year = 24*365.2524
years_per_dataset = n_samples_h/(hours_per_year)

f_per_year = f_per_dataset/years_per_dataset
plt.step(f_per_year, np.abs(fft))
plt.xscale('log')
plt.xlim([0.1, max(plt.xlim())])
plt.xticks([1, 365.2524], labels=['1/Year', '1/day'])
_ = plt.xlabel('Frequency (log scale)')
```

```python
def pre_p(data): 
    data = data.reset_index()
    data=data.drop(['Date'], axis = 1)
    data.drop(data.columns[[2,4,5]], axis=1, inplace=True) 
    return data
```

```python
GOOGL = pre_p(GOOGL)
ADBE = pre_p(ADBE)
DOX = pre_p(DOX)
```

## Norming/Scaling

```python
#just dividing by highest price
def standard_scaler(data):
    scaler = preprocessing.StandardScaler()
    data['High'] = scaler.fit_transform(data.High.values.reshape(-1,1))
    data['Open'] = scaler.fit_transform(data.Open.values.reshape(-1,1))
    data['Close'] = scaler.fit_transform(data.Close.values.reshape(-1,1))
    return data

def min_max_scaler(data):
    scaler = preprocessing.MinMaxScaler()
    data['High'] = scaler.fit_transform(data.High.values.reshape(-1,1))
    data['Open'] = scaler.fit_transform(data.Open.values.reshape(-1,1))
    data['Close'] = scaler.fit_transform(data.Close.values.reshape(-1,1))
    return data
    
def robust_scaler(data):
    scaler = preprocessing.RobustScaler()
    data['High'] = scaler.fit_transform(data.High.values.reshape(-1,1))
    data['Open'] = scaler.fit_transform(data.Open.values.reshape(-1,1))
    data['Close'] = scaler.fit_transform(data.Close.values.reshape(-1,1))
    return data
```

```python
GOOGL = robust_scaler(GOOGL)
ADBE = robust_scaler(ADBE)
DOX = robust_scaler(DOX)
```

```python
window = 7
GX_train, Gy_train, GX_test, Gy_test = split_data(GOOGL, window)
AX_train, Ay_train, AX_test, Ay_test = split_data(ADBE, window)
DX_train, Dy_train, DX_test, Dy_test = split_data(DOX, window)
```

```python
model = make_model([3,window,1])
```

```python
train = [[GX_train, Gy_train], [AX_train, Ay_train], [DX_train, Dy_train]]

for i in train:
    model.fit(
        i[0],
        i[1],
        batch_size=500,
        epochs=500, #nb_epoch
        validation_split=0.1,
        verbose=0) 
```

```python
GtrainScore = model.evaluate(GX_train, Gy_train, verbose=0)
AtrainScore = model.evaluate(AX_train, Ay_train, verbose=0)
DtrainScore = model.evaluate(DX_train, Dy_train, verbose=0)
print('===TRAINING SCORES===')
print('GOOGL: %.5f MSE ' % (GtrainScore[0]))
print('ADBE: %.5f MSE ' % (AtrainScore[0]))
print('DOX: %.5f MSE ' % (DtrainScore[0]))

GtestScore = model.evaluate(GX_test, Gy_test, verbose=0)
AtestScore = model.evaluate(AX_test, Ay_test, verbose=0)
DtestScore = model.evaluate(DX_test, Dy_test, verbose=0)
print('===TEST SCORES===')
print('GOOGL: %.5f MSE ' % (GtestScore[0]))
print('ADBE: %.5f MSE ' % (AtestScore[0]))
print('DOX: %.5f MSE ' % (DtestScore[0]))
```

```python
Gp = model.predict(GX_test)
Ap = model.predict(AX_test)
Dp = model.predict(DX_test)
```
