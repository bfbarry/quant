## Markov Chain

- Obtain gradients (or value) of time series for each time window
    - Need to smooth beforehand because not really differentiable
- Transition mat of one gradient to another
- How to determine good fractional resolution and step size 
    - ratio of unique actions : total actions should be relatively low
round_for_window or window_action first?

```python
def round_for_window(num,frac=None):
    """Round num w.r.t to fractional resolution frac"""
    sign = -1 if num < 0 else 1
    num = np.abs(num)
    increments = np.append(np.arange(0,1,frac),1)
    rounded_prices = [int(num) + i for i in increments]
    diffs = [np.abs(num - r) for r in rounded_prices]
    return sign*rounded_prices[diffs.index(min(diffs))]

def window_action(data,step): # watch out for np.arange float precision
    prices_stepped = [data[t] for t in np.arange(0,len(data),step)]
    return np.diff(prices_stepped)

def transition_series(series, f, s): 
    vals = [round_for_window(p, f) for p in window_action(series,s)]
    for ix,i in enumerate(vals):
        vals[ix] = round(i,2) # undo numpy float weirdness
        if i == 0: vals[ix] = 0 #undo weird -0.0
    return vals
s = 4
pr=prices[:100]
plt.plot(window_action(pr,s))
plt.plot(transition_series(pr,None,s))
```

```python
def transition_mat(series):
    '''Series is a windowed and rounded version of price action'''
    unique = set(series)
    un_ind = {} #indices for each unique action
    for a in unique:
        un_ind[a] = [i for i, x in enumerate(series) if x == a] #for some reason having indices is faster?

    trans = pd.DataFrame(0, columns = unique, index = unique)
    for (a1,a2) in PNR(unique,2): #avoid double loop
        for i in un_ind[a1]:
            try: 
                if series[i+1] == a2: trans.loc[a1][a2] += 1
            except: pass
    return trans.div(trans.sum(axis=1),axis=0).fillna(0)      

# from https://www.analyticsvidhya.com/blog/2021/04/generating-fantasy-titles-with-a-markov-chain/
def markov_sequence(trans, start, t,thresh=0):
    trans = trans.copy(deep=True) #if a copy isn't made it will permanently change the df
    trans[trans < thresh] = 0
    
    prediction = []
    a1 = start
    cols = list(trans.columns)
    col_i = 0
    rand = np.random.rand()
    while len(prediction) != t:
        if col_i >= len(cols):
            print(col_i)
        a2 = cols[col_i]
        p = trans.loc[a1,a2]
        if rand <= p:
            prediction.append(a1)
            a1 = a2
            col_i = 0
            rand = np.random.rand()
        else:
            rand -= p
            col_i += 1
    return prediction

def reconstruct(seq,start):
    series = [start]
    curr=start
    for a in seq:
        curr += a
        series.append(curr)
    return series
```

```python
stop = 120
ser = transition_series(prices[:stop],.5,3)
transition = transition_mat(ser)
predict = markov_sequence(transition, ser[-1], len(ser) + 1)
predicted_prices = reconstruct(predict,prices[stop])
```

```python
plt.figure(figsize=(9,4))
plt.subplot(1,2,1)
plt.hist(predict,bins=20)
plt.subplot(1,2,2)
g = nx.Graph(transition)
nx.draw(g,with_labels=1)
```

```python
plt.figure(figsize=(8,7))
plt.plot(range(len(prices)),prices)
plt.plot(np.arange(stop,len(prices),3),predicted_prices[:-1],label='predicted')
plt.title('Trained on 55% of history, $0.50 resolution, 3 day window')
plt.legend()
```
